```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('fusion.csv')
```

```python
# Distribution des genres
plt.figure(figsize=(12, 6))
genre_counts = df['genre_principale'].value_counts()
plt.title('Distribution des genres de films')
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Nombre de films')
plt.tight_layout()
plt.show()

# Distribution des saisons de sortie
plt.figure(figsize=(10, 6))
season_counts = df['release_season'].value_counts()
plt.title('Distribution des saisons de sortie')
plt.xlabel('Saison')
plt.ylabel('Nombre de films')
plt.tight_layout()
plt.show()
```
